```python
import pandas as pd
import os
```

```python
# Get the path for the csv student file
student_data_to_load = os.path.join("resources","students_complete.csv")
```

```python
# Create dataframe out of student data
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
# Count student data rows and see if there are missing data
student_data_df.count()
student_data_df.isnull().sum()
```

```python
# Another way of checking to see if there are missing data
student_data_df.notnull().sum()
```

```python
# Put the student names in a list.
student_names = student_data_df["student_name"].tolist()
affected_students = []
prefixes = []
suffixes = []

# If the student name has 3 or more split sections, append them to affected students list
for name in student_names:
    if len(name.split()) >= 3:
        affected_students.append(name)

# If the first split section of the student's name is less than or equal to 4, append to prefixes list
# If the last split section of the student's name is less than or equal to 3, append to suffixes list
for name in affected_students:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])
    
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])

set(prefixes)
set(suffixes)
```

```python
for name in affected_students:
     print(name.strip("Mrs. ")) # In this case, the code is removing any of the characters in "Mrs. "
    
```

```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")

student_names = student_data_df["student_name"].tolist()
student_names
```

```python
students_fixed = []

for name in student_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)

len(students_fixed)
```
